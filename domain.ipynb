**Aim**

To perform a basic DNS domain lookup using Python's built-in socket library to resolve domain names to IP addresses without relying on third-party packages.

**Algorithm**

Start

Import Module: Import the built-in socket library.

Input Domain: Define or prompt for the target domain name (e.g., google.com).

Perform Lookup: Use socket.gethostbyname_ex() to query the operating system's native DNS resolver for host name and IP mapping.

Error Handling: Catch socket.gaierror to handle invalid domains or connectivity issues smoothly.

Display Output: Print the primary host name, any aliases, and all resolved IP addresses.

End

```python
import socket

def basic_dns_lookup(target_domain):
    print(f"[+] Performing standard DNS lookup for: {target_domain}\n")
    try:
        # Resolves domain to primary hostname, aliases, and IP list
        hostname, aliases, ip_list = socket.gethostbyname_ex(target_domain)

        print(f"Canonical Hostname : {hostname}")
        print(f"Aliases            : {aliases if aliases else 'None'}")
        print("Resolved IP List   :")
        for ip in ip_list:
            print(f"  [>] {ip}")

    except socket.gaierror:
        print(f"[!] Error: Could not resolve domain '{target_domain}'. Check spelling or connection.")
    except Exception as e:
        print(f"[!] An error occurred: {e}")

if __name__ == "__main__":
    domain = input("Enter target domain: ").strip()
    if domain:
        basic_dns_lookup(domain)
```

**RESULT**

The execution of the DNS lookup script queries public Domain Name System servers to retrieve essential network mapping data for the target domain is successful
