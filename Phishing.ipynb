**Aim**

To develop a Python script that analyzes email headers and content (sender, subject, and body) to flag suspicious patterns such as spoofed domain extensions, urgent or threatening language, and suspicious links, outputting a risk score to determine if an email is phishing.

**Algorithm**

Start

Define Indicator Rules: Establish lists of high-risk sender domains/keywords, urgent subject line triggers, and suspicious body keywords/URLs.

Receive Input: Accept the sender address, subject line, and body text of an incoming email.

Analyze Sender: Check if the sender address contains mismatched top-level domains, lookalike domains, or suspicious keywords.

Analyze Subject: Check the subject line for high-pressure terms (e.g., "Urgent", "Account Suspended", "Action Required").

Analyze Body: Scan the message text for credential-harvesting phrases (e.g., "verify password", "click here") or HTTP links.

Calculate Score: Increment a phishing score counter for every matched indicator.

Classify & End: Compare the total score against a defined threshold to classify the email as "Safe" or "Potential Phishing."

```python
def analyze_phishing(sender, subject, body):
    score = 0
    flags = []

    # Indicator lists
    suspicious_senders = ["@free-mail-service.com", "support-security", "@verify-bank"]
    urgent_subject_words = ["urgent", "suspended", "immediate action", "verify your account", "unauthorized"]
    phishing_body_keywords = ["click here", "update payment", "social security", "password reset", "http://"]


    for indicator in suspicious_senders:
        if indicator in sender.lower():
            score += 2
            flags.append(f"Suspicious sender pattern detected: '{indicator}'")

    for word in urgent_subject_words:
        if word in subject.lower():
            score += 1
            flags.append(f"Urgent keyword in subject: '{word}'")


    for keyword in phishing_body_keywords:
        if keyword in body.lower():
            score += 1
            flags.append(f"Phishing keyword/URL in body: '{keyword}'")


    print("=== PHISHING ANALYSIS REPORT ===")
    print(f"Sender  : {sender}")
    print(f"Subject : {subject}")
    print(f"Score   : {score}\n")

    if flags:
        print("Detected Indicators:")
        for flag in flags:
            print(f"  [!] {flag}")
        print()


    if score >= 3:
        print("[RESULT] HIGH RISK: This email is classified as PHISHING.")
    elif score > 0:
        print("[RESULT] MODERATE RISK: Exercise caution with this email.")
    else:
        print("[RESULT] LOW RISK: No obvious phishing indicators found.")

if __name__ == "__main__":
    sample_sender = "alert@support-security-check.com"
    sample_subject = "URGENT: Your account has been suspended!"
    sample_body = "Dear Customer, Please click here to update payment details immediately."

    analyze_phishing(sample_sender, sample_subject, sample_body)
```

** Result**

The phishing detector evaluates input email components against known social engineering heuristics and suspicious patterns. By calculating a cumulative risk score based on sender domain spoofing, urgency cues, and malicious call-to-action links, the script accurately flags high-risk phishing attempts to prevent credential theft and email-based attacks.
