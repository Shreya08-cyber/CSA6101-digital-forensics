**AIM**

To write and execute a Python program that demonstrates how operating systems automatically update a file's metadata—specifically its Last Access Time (st_atime)—when the contents of that file are read.

**ALGORITHM**

Start


Import Libraries: Load os (for file metadata) and time (for timestamps and delay).

Create File: Create crime_scene.txt and write "Datas collected from the crime scene" into it.

Record Initial Access Time: Fetch file properties using os.stat(), store st_atime in time_before, and print it formatted with time.ctime().

Delay Execution: Pause program execution for 2 seconds using time.sleep(2).

Read File: Open crime_scene.txt and execute file.read() to trigger an OS access event.

Record New Access Time: Fetch file properties again, store updated st_atime in time_after, and print it.

Verify & Display Result: Check if time_before != time_after. If True, print success message confirming the timestamp updated.

Clean Up: Delete crime_scene.txt using os.remove().

End





```python
import os
import time


file_name = "crime_scene.txt"
with open(file_name, "w") as file:
    file.write("Datas collected from the crime scene")


file_info_before = os.stat(file_name)
time_before = file_info_before.st_atime
print("Original Access Time:", time.ctime(time_before))


time.sleep(2)

with open(file_name, "r") as file:
    file.read()


file_info_after = os.stat(file_name)
time_after = file_info_after.st_atime
print("New Access Time:     ", time.ctime(time_after))


if time_before != time_after:
    print("SUCCESS: The computer caught us looking! The time stamp changed.")


os.remove(file_name)
```

**RESULT**

Creation vs. Access: When the file is created, its initial access time (st_atime) is recorded.

Timestamp Shift: Calling file.read() causes the operating system kernel to register a read operation on the file's inode/file attributes, updating the access timestamp to the current clock time.

Verification: The comparison time_before != time_after evaluates to True, successfully demonstrating that file inspection leaves a digital footprint in system metadata.
