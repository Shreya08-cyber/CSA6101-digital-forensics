**Aim**

To write a Python program that generates cryptographic hashes (SHA-256) for digital evidence files to establish and verify their chain of custody. This ensures that digital evidence remains untampered with from the time of acquisition through investigatio

**Algorithm**

Start

Import Module: Import Python's built-in hashlib library.

Define Hash Function: Create a function generate_file_hash(filepath) that reads a file in binary mode ("rb") in fixed chunk sizes (e.g., 4096 bytes) to efficiently process files of any size.

Compute SHA-256 Hash: Update the SHA-256 object with each chunk and return the hexadecimal digest.

Simulate Evidence Acquisition: Calculate and record the initial hash value of an evidence file.

Simulate Verification Step: Recalculate the hash of the file at a later point during the investigation.

Compare Hashes: Match the new hash against the original baseline hash.

Output Result & End: Display whether the integrity is intact (hashes match) or compromised (hashes differ).

```python
import hashlib

def generate_file_hash(filepath):
    """Calculates the SHA-256 hash of a file."""
    sha256_hash = hashlib.sha256()
    try:
        with open(filepath, "rb") as f:
            # Read file in 4KB chunks to handle large files efficiently
            for byte_block in iter(lambda: f.read(4096), b""):
                sha256_hash.update(byte_block)
        return sha256_hash.hexdigest()
    except FileNotFoundError:
        return None

def verify_chain_of_custody(original_hash, filepath):
    """Verifies evidence integrity by comparing current hash with original hash."""
    current_hash = generate_file_hash(filepath)

    print(f"Original Evidence Hash : {original_hash}")
    print(f"Current Evidence Hash  : {current_hash}\n")

    if current_hash is None:
        print("[!] Error: File not found.")
    elif original_hash == current_hash:
        print("[+] VERIFICATION SUCCESSFUL: Chain of custody intact. Evidence is untampered.")
    else:
        print("[!] INTEGRITY ALERT: Evidence hash changed! File has been altered.")

# --- Demonstration ---
if __name__ == "__main__":
    evidence_filename = "evidence.txt"

    # Create sample evidence file
    with open(evidence_filename, "w") as f:
        f.write("Disk dump evidence log data.")

    # 1. Acquire baseline hash
    acquired_hash = generate_file_hash(evidence_filename)

    # 2. Verify evidence integrity
    verify_chain_of_custody(acquired_hash, evidence_filename)
```

**Result**

The program successfully computes the SHA-256 cryptographic checksum of the specified evidence file. By comparing the recalculation against the original recorded hash, the script reliably detects any alteration in file contents down to a single byte, thereby mathematically proving file integrity for legal and forensic chain of custody standards.
